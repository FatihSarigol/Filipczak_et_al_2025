# Compute cluster enrichments and correlations

```python
import pandas as pd

conditions = ['WT', 'KO']
binsizes = [5000]

embeddings = {}
for condition in conditions:
    condition_data = {}
    for binsize in binsizes:
        key = f'{condition}_{binsize}'
        condition_data[binsize] = pd.read_hdf(
            f'../embeddings_genome.h5',
            key
        )
        
    embeddings[condition] = condition_data

conditions = ['WT', 'KO']
embeddings['WT'][5000]
```

```python
for k, embedding in embeddings.items():
    print(k, embedding[5000].columns)
```

```python
from laminatools import vectorize
import numpy as np


ordered_data_columns = {
    'WT': [
        'rnaseq_WT',
        'atacseq_WT',
        'chipseq_BRG1_WT_16',
        'chipseq_BRG1_WT_20', 
        'chipseq_CHD4_WT_16', 
        'chipseq_CHD4_WT_20',
        'chipseq_LAP2alpha_WT_16', 
        'chipseq_LAP2alpha_WT_20',
        'chipseq_laminA_E1_WT_16', 
        'chipseq_laminA_E1_WT_20',
        'chipseq_laminA_3A6_WT_16', 
        'chipseq_laminA_3A6_WT_20'
    ],
    'KO': [
        'rnaseq_KO',
        'atacseq_KO',
        'chipseq_BRG1_KO_16',
        'chipseq_BRG1_KO_20', 
        'chipseq_CHD4_KO_16', 
        'chipseq_CHD4_KO_20',
        'chipseq_LAP2alpha_WT_16', 
        'chipseq_LAP2alpha_WT_20',
        'chipseq_laminA_E1_KO_16', 
        'chipseq_laminA_E1_KO_20', 
        'chipseq_laminA_3A6_KO_16', 
        'chipseq_laminA_3A6_KO_20'
    ]
}


correlations = {}
enrichments = {}
for condition in conditions:
    condition_correlations = {}
    condition_enrichments = {}
    for binsize in binsizes:
        print(condition, binsize)
        df = embeddings[condition][binsize]
        data_columns = np.array(ordered_data_columns[condition])
        cluster_columns = df.columns[
            ~df.columns.isin(['UMAP1', 'UMAP2']) &
            ~df.columns.isin(data_columns) |
            df.columns.str.startswith('louvain') &
            df.columns.str.startswith('chromhmm')
        ]
        df.loc[:, data_columns] = np.log1p(df[data_columns])
        condition_correlations[binsize] = vectorize.data_correlation_over_clusterings(
            df,
            cluster_columns,
            data_columns
        )
        condition_enrichments[binsize] = vectorize.data_enrichment_over_clusterings(
            df,
            cluster_columns,
            data_columns,
            log_results = False
        )

    correlations[condition] = condition_correlations
    enrichments[condition] = condition_enrichments

correlations['WT'][5000]
```

# Compute cluster correspondences between WT and KO

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from laminatools import align, plot

np.random.seed(1094)


for binsize in binsizes:
    for clustering_label, cluster_dict_wt in enrichments['WT'][binsize].items():
        print(binsize, clustering_label)
        cluster_dict_ko = enrichments['KO'][binsize][clustering_label]

        matrix_wt = cluster_dict_wt['matrix']
        cluster_labels_wt = np.array(cluster_dict_wt['row_names'])
        data_labels_wt = cluster_dict_wt['col_names']
        matrix_ko = cluster_dict_ko['matrix']
        cluster_labels_ko = np.array(cluster_dict_ko['row_names'])
        data_labels_ko = cluster_dict_ko['col_names']

        p1, p2, graph, correspondences = align.nearest_neighbour_correspondence(
            matrix_wt, 
            matrix_ko
        )

        fig = plt.figure()
        gs = gridspec.GridSpec(
            len(correspondences) + 1, 2,
            height_ratios = [5] + [1] * len(correspondences)
        )

        ax1 = fig.add_subplot(gs[0, 0])
        ax2 = fig.add_subplot(gs[0, 1])
        plot.make_diagnostic_plot(
            p1, p2, graph, 
            correspondences, 
            cluster_labels_wt, 
            cluster_labels_ko, 
            'WT', 'KO',
            (ax1, ax2)
        )
        ax1.set_title(f'{clustering_label}_conditions')
        ax2.set_title(f'{clustering_label}_correspondences')

        plot.plot_recluster_data(
            correspondences,
            matrix_wt,
            matrix_ko,
            cluster_labels_wt,
            cluster_labels_ko,
            'WT', 'KO',
            gs,
            fig
        )

        fig.set_figwidth(10)
        fig.set_figheight(5 + len(correspondences))
        fig.tight_layout()
        fig.savefig(f'../plots/genome_clustercorrespondence_diagnostic_{clustering_label}_{binsize}.pdf')
        if binsize != 5000:
            plt.close(fig)
```

# Align clusters and datacolumns between conditions

```python
conditions = ['WT', 'KO']
embeddings_new = {condition: {} for condition in conditions}
for binsize in [5000]: #binsizes:
    embedding_wt = embeddings['WT'][binsize].copy()
    embedding_ko = embeddings['KO'][binsize].copy()
    for clustering_label, cluster_dict_wt in enrichments['WT'][binsize].items():
        print(clustering_label)
        cluster_dict_ko = enrichments['KO'][binsize][clustering_label]

        matrix_wt = cluster_dict_wt['matrix']
        cluster_labels_wt = np.array(cluster_dict_wt['row_names'])
        data_labels_wt = cluster_dict_wt['col_names']

        matrix_ko = cluster_dict_ko['matrix']
        cluster_labels_ko = np.array(cluster_dict_ko['row_names'])
        data_labels_ko = cluster_dict_ko['col_names']

        p1, p2, graph, correspondences = align.nearest_neighbour_correspondence(
            matrix_wt, 
            matrix_ko
        )

        for new_cluster_label, correspondence in correspondences.items():
            wt_idx = correspondence['p1_idx']
            ko_idx = correspondence['p2_idx']

            for embedding, old_cluster_labels in zip(
                [embedding_wt, embedding_ko], 
                [cluster_labels_wt[wt_idx], cluster_labels_ko[ko_idx]]
            ):
                embedding.loc[embedding[clustering_label].isin(old_cluster_labels), clustering_label] = 'c' + str(new_cluster_label)
    
    for condition, embedding in zip(conditions, [embedding_wt, embedding_ko]):
        embeddings_new[condition][binsize] = embedding
```

```python
from laminatools import vectorize

# recompute enrichments and correlations
correlations_new = {}
enrichments_new = {}
for condition in conditions:
    condition_correlations = {}
    condition_enrichments = {}
    for binsize in [5000]: #binsizes:
        print(condition, binsize)
        # don't need to log again since we already did above
        df = embeddings_new[condition][binsize]
        
        data_columns = np.array(ordered_data_columns[condition])
        cluster_columns = df.columns[
            ~df.columns.isin(['UMAP1', 'UMAP2']) &
            ~df.columns.isin(data_columns) |
            df.columns.isin(['louvain']) &
            df.columns.str.startswith('chromhmm')
        ]
        condition_correlations[binsize] = vectorize.data_correlation_over_clusterings(
            df,
            cluster_columns,
            data_columns
        )
        condition_enrichments[binsize] = vectorize.data_enrichment_over_clusterings(
            df,
            cluster_columns,
            data_columns
        )

    correlations_new[condition] = condition_correlations
    enrichments_new[condition] = condition_enrichments

correlations_new['WT'][5000]
```

```python
from copy import deepcopy

binsize = 5000
fig, axs = plt.subplots(
    len(enrichments_new['WT'][binsize]),
    len(enrichments_new)
)
for col_axs, (condition, condition_enrichments) in zip(
    axs.T, 
    enrichments_new.items()
):
    for ax, (cluster_column, signal_enrichment) in zip(
        col_axs, 
        condition_enrichments[binsize].items()
    ):
        plot.plot_matrix(
            signal_enrichment['matrix'], 
            signal_enrichment['row_names'], 
            ordered_data_columns['WT'], 
            ax = ax, 
            vmin = -2,
            vmax = 2
        )
        ax.set_ylabel(f'{condition}_{cluster_column}')
        
fig.set_figwidth(10)
fig.set_figheight(5 * len(enrichments_new['WT'][binsize]))
fig.tight_layout()
```

```python
sort_indexes = {
    'louvain_0.1': list(range(23)),
    'chromhmm_5': [3, 2, 0, 1],
    'chromhmm_6': [5, 4, 3, 2, 0, 1],
    'chromhmm_7': [0, 2, 1, 3, 4],
    'chromhmm_8': [5, 3, 2, 4, 0, 1],
    'chromhmm_9': [6, 4, 3, 5, 2, 0, 1],
    'chromhmm_10': [1, 5, 4, 6, 3, 0, 2]
}

binsize = 5000
fig, axs = plt.subplots(
    len(enrichments_new['WT'][binsize]),
    len(enrichments_new)
)
for col_axs, (condition, condition_enrichments) in zip(
    axs.T, 
    enrichments_new.items()
):
    for ax, (cluster_column, signal_enrichment) in zip(
        col_axs, 
        condition_enrichments[binsize].items()
    ):
        print(condition, cluster_column)
        sort_index = sort_indexes[cluster_column]
        plot.plot_matrix(
            signal_enrichment['matrix'][sort_index, :], 
            np.array(signal_enrichment['row_names'])[sort_index], 
            ordered_data_columns['WT'], 
            ax = ax, 
            vmin = -2,
            vmax = 2
        )
        ax.set_ylabel(f'{condition}_{cluster_column}')
        
fig.set_figwidth(10)
fig.set_figheight(5 * len(enrichments_new['WT'][binsize]))
fig.tight_layout()
```

```python
enrichments_sorted = deepcopy(enrichments_new)
correlations_sorted = deepcopy(correlations_new)
embeddings_sorted = deepcopy(embeddings_new)

binsize = 5000
for col_axs, (condition, condition_enrichments) in zip(
    axs.T, 
    enrichments_new.items()
):
    for ax, (cluster_column, signal_enrichment) in zip(
        col_axs, 
        condition_enrichments[binsize].items()
    ):
        print(cluster_column, condition, binsize)
        sort_index = sort_indexes[cluster_column]
        embedding = embeddings_sorted[condition][binsize]
        row_labels = np.array(
            enrichments_sorted[condition][binsize][cluster_column]['row_names']
        )
        sorted_row_labels = row_labels[sort_index]
        
        for signals in [enrichments_sorted, correlations_sorted]:
            signal = signals[condition][binsize][cluster_column]
            signal['matrix'] = signal['matrix'][sort_index, :]
            signal['row_names'] = ['c' + str(i) for i in range(len(sorted_row_labels))]
        
        old_to_new_cluster_names = {
            old_row_name: f'c{i}' for i, old_row_name in enumerate(sorted_row_labels)
        }
        embedding.loc[:, cluster_column] = embedding.loc[:, cluster_column].apply(
            lambda x: old_to_new_cluster_names[x]
        )
```

# Generate correspondence plots

```python
from laminatools import plot, enrichment

np.random.seed(1094)

    
fdr = 0.15
lap2_enriched_bins = {}
this_condition, that_condition = conditions
palettes = {}
binsize = 5000

binsize_correlations_this = correlations_sorted[this_condition][binsize]
binsize_correlations_that = correlations_sorted[that_condition][binsize]
binsize_enrichments_this = enrichments_sorted[this_condition][binsize]
binsize_enrichments_that = enrichments_sorted[that_condition][binsize]

this_data = embeddings_sorted[this_condition][binsize]
that_data = embeddings_sorted[that_condition][binsize]

lap2_columns = this_data.columns[this_data.columns.str.contains('LAP2')]
enriched_bins = set()
for lap2_column in lap2_columns:
    lap2_enriched_bins = enrichment.get_enriched_bins(
        this_data[lap2_column],
        fdr
    )
    print(
        len(lap2_enriched_bins), 
        'of', 
        len(this_data), 
        'bins are enriched in', 
        lap2_column
    )
    enriched_bins.update(lap2_enriched_bins)
    
for data in [this_data, that_data]:
    data['lap2_enriched'] = [i in enriched_bins for i in data.index]

for clustering in binsize_correlations_this.keys():
    colors = plot.get_categorical_colors(
        binsize_enrichments_this[clustering]['row_names']
    )
    palettes[clustering] = colors
    palette = {
        f'{i+1}_{condition}': colors for i, condition in enumerate(conditions)
    }
    plot.plot_cluster_correspondence(
        binsize_correlations_this,
        binsize_correlations_that,
        this_data,
        that_data,
        clustering,
        this_condition,
        that_condition,
        f'../plots/clustercorrespondence_correlation_{binsize}_{clustering}_all.pdf',
        vmin = -1,
        vmax = 1,
        palette = palette
    )

    plot.plot_cluster_correspondence(
        binsize_correlations_this,
        binsize_correlations_that,
        this_data,
        that_data,
        clustering,
        this_condition,
        that_condition,
        f'../plots/clustercorrespondence_correlation_{binsize}_{clustering}_lap2.pdf',
        vmin = -1,
        vmax = 1,
        subset = list(enriched_bins),
        palette = palette
    )

    plot.plot_cluster_correspondence(
        binsize_enrichments_this,
        binsize_enrichments_that,
        this_data,
        that_data,
        clustering,
        this_condition,
        that_condition,
        f'../plots/clustercorrespondence_enrichment_{binsize}_{clustering}_all.pdf',
        vmin = -2,
        vmax = 2,
        palette = palette
    )

    plot.plot_cluster_correspondence(
        binsize_enrichments_this,
        binsize_enrichments_that,
        this_data,
        that_data,
        clustering,
        this_condition,
        that_condition,
        f'../plots/clustercorrespondence_enrichment_{binsize}_{clustering}_lap2.pdf',
        vmin = -2,
        vmax = 2,
        subset = list(enriched_bins),
        palette = palette
    )
```
